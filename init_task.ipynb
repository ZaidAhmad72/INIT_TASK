##C4. Customer Segmentation (RFM)


```python
full_data.columns
```

```python
full_data.head(3)
```

Using Recency, Frequency, Monetary (RFM) logic:

1.Recency: Days since last purchase (from latest transaction date) <br>
2.Frequency: Total number of transactions <br>
3.Monetary: Total revenue generated


```python
latest_transaction_date = full_data['timestamp'].max()
```

```python
def days_since_lastPurchase(series):
    return (latest_transaction_date - series.max()).days
```

```python
rfm_data = full_data.groupby('customer_id').agg(
    Recency=('timestamp', days_since_lastPurchase),
    Frequency=('transaction_id', 'count'),
    Monetary=('revenue', 'sum')
).reset_index()
```

#Part D: Advanced Operations

##Efficient Operations Challenge

Identifying suspicious transactions(quantity > 100 AND price < 10,
 OR Customers with more than 3 purchases in a single hour.)

```python
def find_suspicious_transactions(full_data):
    #  Identifying suspicious transactions using vectorized operations

    mask_price_qty = (full_data['quantity'] > 100) & (full_data['price'] < 10)

    # Group by customer_id, day, and hour, then count transactions
    hour_count = full_data.groupby(['customer_id', full_data['timestamp'].dt.date, full_data['transaction_hour']])[
        'transaction_id'
    ].transform('count')

    mask_frequency = hour_count > 3

    #Merge both the suspicions
    suspicious_mask = mask_price_qty | mask_frequency

    suspicious_df = full_data[suspicious_mask].copy()

    #Create reason column

    conditions = [mask_price_qty[suspicious_mask], mask_frequency[suspicious_mask]]
    reasons = [
        "Unusual Qty/Price combination",
        "More than 3 purchases"
    ]

    # np.select applies the reasons based on the conditions
    suspicious_df['reason'] = np.select(conditions, reasons, default='Multiple reasons')

    return suspicious_df[['transaction_id', 'customer_name', 'reason']]
```

```python
find_suspicious_transactions(full_data)
```
